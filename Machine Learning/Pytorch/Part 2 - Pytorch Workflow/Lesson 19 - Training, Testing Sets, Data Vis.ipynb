### Spliting Data Into Training And Testing Set
(one of the most important concepts in machine learning in general)

There are 3 Datasets:
1. Course materials - training set
2. Practice Exam - validation set used to tune the models patterns
3. Final Exam - test set used to see if the model is ready to be used in the real world

The whole point of this is to see how well the model performs on data that it hasn't seen yet

Generalization: The abality for a machine learning mode to perform well on data that it hasn't seen before (has the model learned any skills from the materials, are you ready to go into the real world)
- making decisions based on the new data
- from the patterns learned in the dataset

```python
import torch
import matplotlib.pyplot as plt
```

```python
weight = 0.91
bias = 0.3

# Create
start = 0
end = 1
step = 0.02
# parameters for creating the tensor: X

X = torch.arange(start, end, step).unsqueeze(dim = 1)
# adds an extra dimension so theres an extra square bracket and each element of X, y are on different lines in the output for betting viewing
y = weight * X + bias # is Linear Regression Formula

# code imported from lesson 18
```

```python
# Create a train/test split

trainSplit = int(0.8 * len(X)) # creating the train split by multiplying the upper bounds of the train split by the length of X to get the total number

XTrain, yTrain = X[:trainSplit], y[:trainSplit] # indexing to get all samples up until the trainsplit
XTest, yTest = X[trainSplit:], y[trainSplit:] # indexing to get all the samples from the trainsplit onwards, or what is left over after the trainsplit

len(XTrain), len(yTrain), len(XTest), len(yTest)
```

### Building A Function To Visualize Our Data

Data Explorers Motto: Whenever you don't understand a concept, visualize the concept

```python
# using matplotlib to visualize the data points

def plotPredictions(trainData = XTrain, 
                    trainLabels = yTrain, 
                    testData = XTest, 
                    testLabels = yTest, 
                    prediction = None):
# Plots training data, test data and comparing predictions

    plt.figure(figsize = (10, 7)) 
    
    plt.scatter(trainData, trainLabels, c = "b", label = "Training Data")

    plt.scatter(testData, testLabels, c = "g", label = "Testing Data")

    if prediction is not None:
        plt.scatter(testData, prediction, c = "r", label = "Prediction")

    plt.legend(prop = {"size": 14})
```

```python
plotPredictions()

# blue dots are the inputs, and the green dots are the ideal/perfect output
# only if the model is perfect will the model output the exact green dots
```
