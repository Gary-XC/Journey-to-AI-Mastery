#### Introductions To Tensors

##### Creating Tensors
Pytorch Tensors are created using torch.tensor() = https://pytorch.org/docs/stable/tensors.html

```python
# scalar tensor

scalar = torch.tensor(7)
scalar
```

```python
scalar.ndim
# .ndim to find the number of dimensions
# but snce it is a scalar we know that the there are no dimensions
```

```python
# using .item() we can get back the element of the tensor as a Python integer
scalar.item()
```

```python
# a vector is magnitude and direction, and has more then one number
vector = torch.tensor([7,7])
vector
```

```python
vector.ndim
# the number of dimenstions that a tensor has is based on the number of arrays in the tensor
```

```python
vector.shape
# Shape is finding out the number of pairs of closed square brackets
# 2 by 1 elements
```

```python
# MATRIX
MATRIX = torch.tensor([[7,8], 
                       [9, 10]])
MATRIX
```

```python
MATRIX.ndim
# has 2 dimensions as there are 2 arrays in the tensor
```

```python
MATRIX[1]
# works the same way that multiple dimenstion arrays would work in normal python and java
```

```python
MATRIX.shape
# is 2x2 as there are 2 arrays that each have 2 elements in them
```

```python
# TENSOR
TENSOR = torch.tensor([[[1, 2, 3],
                        [3, 6, 9],
                        [2, 4, 6]]])

TENSOR
```

```python
TENSOR.ndim
```

```python
TENSOR.shape
# the first dimension is also considered the 0th dimension
```

```python
TENSOR[0]
# when trying to call elements from tensors the first element is always the 0th element?
# and therefore must always be called first before getting to the other rows, columns, or elements
```

```python
TENSOR[0][2][1]
# first index is calling the 0th dimension, and is needed for the logic of the tensor
# the second index is calling the second row of the index
# and finally the third index is calling the second element of the second row
```
